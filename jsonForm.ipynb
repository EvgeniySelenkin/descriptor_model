```python
import json
```

```python
topics_artm = []
with open('data/topics/topics_artm_dict_2.txt', 'r', encoding='utf-8') as read_file:
    topics_artm = json.load(read_file)
```

```python
topics_artm
```

```python
json_artm = {
    "name": "Дескрипторная моель 2",
    "children": [
        {
            "name": "Знать",
            "children": []
        },
        {
            "name": "Уметь",
            "children": []
        },
        {
            "name": "Владеть",
            "children": []
        }
        
    ]
    }
```

```python
topics_know = []
for pair in topics_artm[0].items():
    topic = {"name": pair[0], "children": []}
    terms = []
    for elem in pair[1].split(', ')[:-1]:
        word = elem.split('-')
        term = {"name": word[0], "weight": float(word[1])}
        terms.append(term)
    topic["children"] = terms
    topics_know.append(topic)
```

```python
json_artm["children"][0]["children"] = topics_know
```

```python
topics_can = []
for pair in topics_artm[1].items():
    topic = {"name": pair[0], "children": []}
    terms = []
    for elem in pair[1].split(', ')[:-1]:
        word = elem.split('-')
        term = {"name": word[0], "weight": float(word[1])}
        terms.append(term)
    topic["children"] = terms
    topics_can.append(topic)

json_artm["children"][1]["children"] = topics_can
```

```python
topics_skill = []
for pair in topics_artm[2].items():
    topic = {"name": pair[0], "children": []}
    terms = []
    for elem in pair[1].split(', ')[:-1]:
        word = elem.split('-')
        term = {"name": word[0], "weight": float(word[1])}
        terms.append(term)
    topic["children"] = terms
    topics_skill.append(topic)

json_artm["children"][2]["children"] = topics_skill
```

```python
with open('models/model_2.json', 'w', encoding='utf-8') as write_file:
  write_file.write(str(json_artm))
```
