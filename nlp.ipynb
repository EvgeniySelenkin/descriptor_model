```python
import fitz
import json
import re
import nltk
from pymystem3 import Mystem
import nltk
import string
from nltk.corpus import stopwords
import concurrent.futures
```

```python
docs = ['1.pdf', '2.pdf', '3.pdf', '4.pdf']
documents = []
for pdf in docs:
    doc = fitz.open(pdf)
    document = ""
    for i in range(doc.pageCount):
        page = doc.load_page(i)
        document += page.get_text('text')
    documents.append(document)
```

```python
def remove_garbage(raw_text):
  text = re.sub('\W', ' ', raw_text)
  text = re.sub('\d', ' ', text)
  return text

print(remove_garbage('wertyu8585.,'))
```

```python
nltk.download('punkt')
from nltk.tokenize import word_tokenize
def tokenize(raw_text):
  tokens = word_tokenize(raw_text)
  return tokens
```

```python
nltk.download('wordnet')
```

```python
def to_base_form(raw_text):
  mstem = Mystem()
  tokens = mstem.lemmatize(raw_text)
  string = ''
  for elem in tokens[:-1]:
    string = string + elem
  return tokens
```

```python
nltk.download('stopwords')
```

```python
stop_words = set(stopwords.words('russian'))
punct = string.punctuation
def remove_stop_words(raw_text):
  text = [word for word in raw_text if not word in stop_words]
  text = [word for word in raw_text if not word in punct]
  return text
```

```python
def get_prepared(document):
  document_without = remove_garbage(document)
  base_document = to_base_form(document_without)
  #document_sentences = re.split(r'(?<=\w[.!?]) ', base_document)
  #prepared_document = []
  '''for sentence in document_sentences:
    sentence_without = remove_garbage(sentence)
    tokenize_sentences = tokenize(sentence_without)
    correct_text = remove_stop_words(tokenize_sentences)
    prepared_document.append(correct_text)
    print('s')'''
  print('D')
  return base_document
```

```python
def get_prepared_all_documents(documents):
    result = []
    for document in documents:
        result.append(get_prepared(document))
    return result
```

```python
result = get_prepared_all_documents(documents)
```

```python
print(range(len(result)))
```

```python
for i in range(len(result)):
    for j in range(len(result[i])):
        result[i][j] = re.sub('\s+', ' ', result[i][j])
print(result)
```

```python
result_without_stopwords = []
for elem in result:
    result_without_stopwords.append(remove_stop_words(elem))
result_without_stopwords
```

```python
docs = []
for elem in result_without_stopwords:
    doc_string = ''
    for e in elem:
        doc_string += e
    docs.append(doc_string)
print(docs)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf_vectorizer = TfidfVectorizer(max_features=100)
tfidf = tfidf_vectorizer.fit_transform(docs)
tfidf_vectorizer.vocabulary_
```
